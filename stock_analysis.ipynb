```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
def retrieve_dataframes(ticker):
    bs=ticker.quarterly_balancesheet
    fi=ticker.quarterly_financials
    cf=ticker.quarterly_cashflow
    earnings=ticker.quarterly_earnings
    history=ticker.history(period='2y')

    bs.insert(0,'Stock',ticker.ticker)
    fi.insert(0,'Stock',ticker.ticker)
    cf.insert(0,'Stock',ticker.ticker)
    earnings.insert(0,'Stock',ticker.ticker)
    history.insert(0,'Stock',ticker.ticker)

    bs=bs.reset_index()
    fi=fi.reset_index()
    cf=cf.reset_index()
    earnings=earnings.reset_index()
    history=history.reset_index()
    return [bs,fi,cf,earnings,history]

def merge(dfset1, dfset2):
    result=[]
    for i in range(len(dfset1)):
        df=pd.concat([dfset1[i],dfset2[i]],axis=0)
        result.append(df)
    return result

def savecsv(dfset):
    (bs,fi,cf,earnings,history)=(dfset[0],dfset[1],dfset[2],dfset[3],dfset[4])
    bs.to_csv("bs.csv")
    fi.to_csv("fi.csv")
    cf.to_csv("cf.csv")
    earnings.to_csv("earnings.csv")
    history.to_csv("history.csv")  
```

```python
mytickers=yf.Tickers("BAC MSFT")
```

```python
all_result=[pd.DataFrame(),pd.DataFrame(),pd.DataFrame(),pd.DataFrame(),pd.DataFrame()]
for myticker in mytickers.tickers:
    all_result = merge(all_result,retrieve_dataframes(myticker))
    
savecsv(all_result)
```
